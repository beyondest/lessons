```python
img=cv2.imread(path)
imo.plt_show0(img)

img=img.reshape((1280,1024,3))
imo.plt_show0(img)
```

```python
img_filter=cv2.GaussianBlur(img,(3,3),1)
#img_filter=cv2.blur(img,(5,5))
imo.plt_show0(img_filter)
```

```python
img_yuv=cv2.cvtColor(img_filter,cv2.COLOR_BGR2YUV)
y,u,v=cv2.split(img_yuv)


#low=np.array([158, 90, 58], dtype="uint8")
#up=np.array([176,128,91],dtype='uint8')
v=cv2.inRange(v,58,91)
imo.plt_show(v)
```

```python
u_bin=cv2.inRange(u,90,128)
imo.plt_show(u_bin)
```

```python
u_filter=cv2.medianBlur(u_bin,13)
imo.plt_show(u_filter)
```

```python
kernel=cv2.getStructuringElement(cv2.MORPH_RECT,(10,10))
u_close=cv2.morphologyEx(u_filter,cv2.MORPH_CLOSE,kernel,iterations=1)
imo.plt_show(u_close)
```

```python
u_bgr=cv2.cvtColor(u_close,cv2.COLOR_GRAY2RGB)

u_cont=imo.make_cont(u_bgr,500)
imo.plt_show0(u_cont)
```
