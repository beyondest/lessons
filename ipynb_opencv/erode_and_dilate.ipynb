```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('../resource/corrosion.jpg')
kernal_3=np.ones((3,3),np.uint8)
img_erosion_1=cv2.erode(img,kernal_3,iterations=1)
img_erosion_2=cv2.erode(img,kernal_3,iterations=2)
img_erosion_3=cv2.erode(img,kernal_3,iterations=3)
res=np.hstack((img,img_erosion_1,img_erosion_2,img_erosion_3))
plt.imshow(res)
```

```python
img_dilate1=cv2.dilate(img,kernal_3,iterations=1)
img_dilate2=cv2.dilate(img,kernal_3,iterations=2)
img_dilate3=cv2.dilate(img,kernal_3,iterations=3)
res2=np.hstack((img,img_dilate1,img_dilate2,img_dilate3))
plt.imshow(res2)
```
