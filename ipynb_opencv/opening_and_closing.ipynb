```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('../resource/corrosion.jpg')
kn1=np.ones((10,10),np.uint8)
kn2=np.ones((20,20),np.uint8)
img_opening=cv2.morphologyEx(img,cv2.MORPH_OPEN,kn1)
ret=np.hstack((img,img_opening))
plt.imshow(ret)
```

```python
img_closing=cv2.morphologyEx(img,cv2.MORPH_CLOSE,kn2)
ret2=np.hstack((img,img_closing))
plt.imshow(ret2)
```
