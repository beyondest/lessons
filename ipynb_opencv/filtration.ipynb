```python
import cv2
import matplotlib.pyplot as plt
img=plt.imread('/home/liyuxuan/vscode/res/lena.jpg')
plt.imshow(img)
```

```python
img_blur=cv2.blur(img,(13,13))
plt.imshow(img_blur)
```

```python
img_box=cv2.boxFilter(img,-1,(3,3),normalize=False)
'''if normalize=ture , same as blur'''
plt.imshow(img_box)
```

```python
img_gaussian=cv2.GaussianBlur(img,(13,13),1)
plt.imshow(img_gaussian)
```

```python
img_median=cv2.medianBlur(img,13)
plt.imshow(img_median)
```

```python
import numpy
res=numpy.hstack((img,img_blur,img_box,img_gaussian,img_median))
plt.imshow(res)
```
