```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread("../resource/corrosion.jpg")
kn1=np.ones((20,20),np.uint8)
img_tophat=cv2.morphologyEx(img,cv2.MORPH_TOPHAT,kn1)
img_blackhat=cv2.morphologyEx(img,cv2.MORPH_BLACKHAT,kn1)
ret=np.hstack((img,img_tophat,img_blackhat))
plt.imshow(ret)
```
