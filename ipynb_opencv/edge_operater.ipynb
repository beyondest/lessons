```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread("../resource/honkai.jpg",cv2.IMREAD_GRAYSCALE)
img_sobelx=cv2.Sobel(img,cv2.CV_64F,1,0,ksize=3)
img_sobely=cv2.Sobel(img,cv2.CV_64F,0,1,ksize=3)
img_sobelx=cv2.convertScaleAbs(img_sobelx)
img_sobely=cv2.convertScaleAbs(img_sobely)
img_sobel=cv2.addWeighted(img_sobelx,0.5,img_sobely,0.5,0)
ret=np.hstack((img,img_sobel,img_sobelx,img_sobely))
plt.imshow(ret,'gray')
```
